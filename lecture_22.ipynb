```python
class MathUtils:
    @staticmethod
    def add(a,b):
        return a + b
    
    @staticmethod
    def multiply(a,b):
        return a * b
```

```python
print(MathUtils.add(5,3))
print(MathUtils.multiply(5,3))
```

```python
class Person:
    def __init__(self, name, age):
        if not Person.is_valid_age(age):
            raise ValueError("Age must be between 0 and 120")
        
        self.name = name
        self.age = age

    @staticmethod
    def is_valid_age(age):
        return 0 <= age <= 120
```

```python
p1 = Person("Ali", 25)
```

```python
p1.name
```

```python
class Person:
    species = "Human"

    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    @classmethod
    def change_species(cls, new_species):
        cls.species = new_species
```

```python
Person.change_species("aliens")
```

```python
class CountDown:
    def __init__(self, start):
        self.current = start

    def __iter__(self):
        return self
    
    def __next__(self):
        if self.current <= 0:
            raise StopIteration
        val = self.current
        self.current -= 1
        return val
```

```python
for num in CountDown(20):
    print(num)
```

```python
def simple_gen():
    yield "Hello"
    yield "World"

for word in simple_gen():
    print(word)
```

```python
file_path = "Docs/test.txt"

def read_file(file_path):
    with open(file_path) as f:
        for line in f:
            yield line.strip()
```

```python
for line in read_file(file_path):
    print(line)
```
