1.6 Loops

for vs while

for is like checking each box in a tray; while is like filling glasses until the jug is empty.

```python
# Let's loop through a string
word = "Python"
print("Let's spell out Python:")

for letter in word:
    print(f"The letter is: {letter}")

print("\nNow you know how to spell Python!")
```

```python
# Let's count some animals in a zoo!
animals = ["elephant", "giraffe", "penguin", "lion", "zebra"]

print("Let's visit each animal in the zoo:")

for animal in animals:
    print(f"I see a {animal}!")

print("\nWe've seen all the animals! What a fun day at the zoo!")

# Let's count from 1 to 5
print("\nNow let's count together:")

for number in range(1, 6):
    print(f"Count with me: {number}!")
```

```python
# Let's count fruits in a basket
fruits = ["apple", "banana", "orange", "mango", "grape"]
print("Checking fruits in my basket:")
for fruit in fruits:
    print(f"I have a fresh {fruit}")

print("\n")

# Let's add numbers in a list
numbers = [1, 2, 3, 4, 5]
total = 0
print("Adding numbers:")
for num in numbers:
    total += num
    print(f"Current sum is: {total}")

print("\n")

# Let's check temperatures throughout the day
temperatures = [72, 75, 80, 78, 71]
times = ["Morning", "Noon", "Afternoon", "Evening", "Night"]
print("Today's temperature readings:")
for i in range(len(temperatures)):
    print(f"{times[i]}: {temperatures[i]}Â°F")

print("\n")

# Let's spell a word backwards
word = "HELLO"
print("Spelling HELLO backwards:")
for char in reversed(word):
    print(char)

print("\n")

# Let's count by twos
print("Counting by twos up to 10:")
for num in range(0, 11, 2):
    print(f"Number: {num}")
```

```python
# Let's simulate a student's weekly study schedule
# This example demonstrates nested for loops

subjects = ["Math", "Science", "History"]
days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday"]

total_study_hours = 0

# Outer loop: Iterate through each day
for day in days:
    print(f"\n=== {day}'s Study Schedule ===")
    
    # Inner loop: Study each subject
    for subject in subjects:
        # Simulate study hours (2-4 hours per subject)
        hours = 3
        print(f"Studied {subject} for {hours} hours")
        total_study_hours += hours

print("\n=== Weekly Summary ===")
print(f"Total hours studied: {total_study_hours}")
print(f"Average hours per day: {total_study_hours/len(days):.1f}")
```

1.7 Break & Continue

Add a rule: If stock runs out, close the stall early.

```python
# Example 1: Break when lemons run out
lemons = 10
for customer in range(20):
    if lemons == 0:
        print("Sorry, we're sold out!")
        break
    print(f"Serving customer {customer}")
    lemons -= 1

print("\n")

# Example 2: Continue if customer has no money
customers = ["Alice", "Bob", "Charlie", "David", "Eve"]
for customer in customers:
    if customer == "Charlie":  # Charlie forgot his wallet
        print(f"Skipping {customer} - no money")
        continue
    print(f"Serving {customer}")

print("\n")

# Example 3: Break when finding first even number
numbers = [7, 3, 9, 4, 1, 6]
for num in numbers:
    if num % 2 == 0:
        print(f"First even number found: {num}")
        break
    print(f"Checking {num}")

print("\n")

# Example 4: Continue for invalid scores
scores = [85, -10, 92, 100, -5, 88]
for score in scores:
    if score < 0:
        print(f"Skipping invalid score: {score}")
        continue
    print(f"Valid score: {score}")

print("\n")

# Example 5: Break when password found
passwords = ["123456", "password", "correct_pass", "qwerty"]
correct_password = "correct_pass"
for attempt in passwords:
    if attempt == correct_password:
        print("Correct password found!")
        break
    print(f"Wrong password: {attempt}")
```
